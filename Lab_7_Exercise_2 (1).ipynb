```
import requests
import json
```

```
url = f"https://date.nager.at/api/v3/CountryInfo/1"   #Calls the country info

country_info_data = requests.get(url)    #this pulls the countrty info data
json_file = country_info_data.json()   #puts the data into a json file

json_file   #prints the data
```

```
url = f"https://date.nager.at/api/v3/AvailableCountries"    #calls available countries

available_countries_data = requests.get(url)
json_file = available_countries_data.json()

json_file
```

```
url = f"https://date.nager.at/api/v3/LongWeekend/2023/62"   #Calls the long weekend with a country and year

long = requests.get(url)
json_file = long.json()

json_file
```

```
url = f"https://date.nager.at/api/v3/PublicHolidays/2023/82"    #Calls the public holiday with the year and country code

public = requests.get(url)
json_file = public.json()

json_file
```

```
url = f"https://date.nager.at/api/v3/IsTodayPublicHoliday/UK?offset=0"    #This call checks to see if today is a public holiday
   #I chose the UK for my country
today = requests.get(url)
json_file = today.json()

json_file
```
